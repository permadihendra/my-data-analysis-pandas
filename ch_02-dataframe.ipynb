### Using Regular Expression - Regex
#### select all the earthquakes in California that have magnitudes of at least 3.8. We need to select entries in the place column that end in CA or California because the data isn't consistent (we will look at how to fix this in the next section). The $ character means end and 'CA$' gives us entries that end in CA, so we can use 'CA|California$' to get entries that end in either

```python
df.loc[
    (df.place.str.contains(r'CA|California$')) & (df.mag > 3.8),
    ['alert','mag','magType','title','tsunami','type', 'place']
].sort_values(by='mag', ascending=False)
```

```python
(df.mag.idxmin(), df.mag.idxmax())
```

```python
df.loc[
    (df.mag.idxmin(), df.mag.idxmax()),
    ['alert','mag','magType','title','tsunami','type', 'place']
].sort_values(by='mag', ascending=False)
```
